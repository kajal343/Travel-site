```python
import pandas as pd
food =pd.read_csv(r"C:\Users\Dell\Downloads\archive (3)\zom_jaipur.csv", encoding='latin1')
food
```

```python
food1 = food.drop_duplicates()
food1 = food1.dropna()
food1 = food1.drop(columns=["Url","din_rev","del_rev","home_delivery","avg_del_rate"])
food1
```

```python

# Function to recommend restaurants based on user preferences
def recommend_restaurants(user_category, is_vegetarian, max_cost, user_cuisine, top_n=10):
    if not pd.api.types.is_numeric_dtype(food1['cost_for_two']):
        # Convert the "cost_for_two" column to float
        food1['cost_for_two'] = food1['cost_for_two'].str.replace(',', '').astype(float)

#     food1['cost_for_two'] = food1['cost_for_two'].str.replace(',', '').astype(float)

   
# # food1['cost_for_two'] = food1['cost_for_two'].str.replace(',', '').astype(float)
     

    # Filter restaurants based on user preferences (category, vegetarian, cuisine, and cost)
    filtered_restaurants = food1[
        (food1['category'].str.contains(user_category, case=False, na=False)) &
        ((food1['is_vegeterian'].str.lower() == 'yes') if is_vegetarian else (food1['is_vegeterian'].str.lower() == 'no')) &
        (food1['cost_for_two'] <= max_cost) &
        (food1['Cuisine_offered'].str.contains(user_cuisine, case=False, na=False)) &  # Note the closing parenthesis here
        (food1["Location"])

    ]

    # Sort restaurants by rating and select the top N
    sorted_restaurants = filtered_restaurants.sort_values(by='avg_din_rate', ascending=False).head(top_n)

    return sorted_restaurants

# User inputs
user_category = input("Enter restaurant category: ").strip()
is_vegetarian = input("Vegetarian (yes/no): ").strip().lower() == 'yes'
max_cost = float(input("Maximum co'st (e.g., 500): "))
user_cuisine = input("Enter desired cuisine: ").strip()

# Get recommended restaurants based on user preferences (top 10)
recommended_restaurants = recommend_restaurants(user_category, is_vegetarian, max_cost, user_cuisine, top_n=10)


if not recommended_restaurants.empty:
    print("Recommended restaurants (Top 10):")
    print()
    print("---------------------------------------------")
    for idx, row in recommended_restaurants.iterrows():
        print(f"Name: {row['Name']}")
        print(f"Location: {row['Location']}")
        print(f"Category: {row['category']}")
        print(f"Bar available: {row['is_bar_available']}")
        print(f"Cost: {row['cost_for_two']}")
        print(f"Rating: {row['avg_din_rate']}")
        
        print()  
else:
    print("No recommendations found for the given preferences.")
```
